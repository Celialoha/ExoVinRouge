```python
import pandas as pd
import numpy as np
```

```python
def redWineDf(df_to_format):
    list_headers = [11,12,13,14,15,16,17,18]
    tab_name_columns = ["Departements","Déclarations","Superficie Totale Prod"
                ,"Superficie A.O.P","Aptes au Cognac ou Armagnac","Superficie IGP","Superficie VSIG","QT AOP Blanc",
                "QT AOP Rouge","QT AOP Rosé", "QT AOP VSI/VCI","QT IGP Blanc","QT IGP Rouge",
               "QT IGP Rosé","QT IGP VCI","QT VSIG Blanc","QT VSIG Rouge","QT VSIG Rosé",
                "Quantité Aptes Cognac/Armagnac","Production Commerciale","Production Non Commerciale","Total"]
    df_to_format.columns = tab_name_columns
    numDepartement = []
    for i in range(len(df_to_format.Departements.values)):
    #If departements numbers starts with a 0, delete it
        if(df_to_format.Departements.values[i][0] == '0'):
            numDepartement.append(df_to_format.Departements.values[i][1])
        else:
            numDepartement.append((df_to_format.Departements.values[i][0]+df_to_format.Departements.values[i][1]))
        df_to_format.Departements.values[i] = df_to_format.Departements.values[i][3::]
    df_to_format.index = numDepartement
    df_to_format.index.name = "Num_Departement"
    df_to_format.rename(columns={'Departements': 'Nom_Departement'}, inplace=True)
    formated_dataframe = df_to_format[["Nom_Departement","Déclarations","Superficie Totale Prod","Superficie A.O.P",
      "Superficie IGP","Superficie VSIG","QT AOP Rouge","QT IGP Rouge","QT VSIG Rouge","Production Commerciale",
      "Production Non Commerciale","Total"]]
    return formated_dataframe
```

```python
vini2015 = pd.read_excel("https://www.data.gouv.fr/fr/datasets/r/98f4a018-de6d-4a5a-9c4a-4170f977626d")
vini2015 = vini2015[19:97]
vini2015 = redWineDf(vini2015)
vini2015
```
